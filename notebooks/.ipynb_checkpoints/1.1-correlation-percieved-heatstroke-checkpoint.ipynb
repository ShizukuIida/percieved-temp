# Data Loading

```python
import datetime
import pandas as pd
```

```python
processed_data_path = '../data/processed/'
cri_1 = datetime.datetime(2022, 6, 1)
cri_2 = datetime.datetime(2022, 9, 30)

#熱中症データ
df_stroke = pd.read_csv(processed_data_path+'heat_stroke.csv')
df_stroke = df_stroke.rename(columns={'日付': 'datetime'})
df_stroke['datetime'] = pd.to_datetime(df_stroke['datetime'])
df_stroke = df_stroke.set_index('datetime')
```

```python
df_stroke.head()
```

```python
#体感温度データ
df_apparent = pd.read_csv(processed_data_path+'apparent_rate_since_05_29_2022.tsv', sep='\t')
df_apparent['体感温度'] = df_apparent['hot_and_cold'] + df_apparent['hot_and_not_cold']
df_apparent = df_apparent[(pd.to_datetime(df_apparent['datetime']) >= cri_1)]
df_apparent = df_apparent[(pd.to_datetime(df_apparent['datetime']) <= cri_2)]
df_apparent = df_apparent.reset_index(drop=True)
df_apparent['datetime'] = pd.to_datetime(df_apparent['datetime'])
df_apparent = df_apparent.set_index('datetime')
```

```python
#気象データ
df_kisyo = pd.read_csv(processed_data_path+'weather_data_06-09.csv')
df_kisyo['MC'] = df_kisyo['平均気温(℃)'] - (df_kisyo['平均気温(℃)'] - 10)*(0.8 - df_kisyo['平均湿度(％)']/100)/2.3
df_kisyo['NET'] = 37 - ((37 - df_kisyo['平均気温(℃)'])/( 0.68 - 0.0014*df_kisyo['平均湿度(％)'] + 1 /(1.76+1.4*(df_kisyo['平均風速(m/s)'])**0.75)))-0.29*df_kisyo['平均気温(℃)']*(1-df_kisyo['平均湿度(％)']/100)
df_kisyo['THI'] = 0.81*df_kisyo['平均気温(℃)'] + 0.01*df_kisyo['平均湿度(％)']*(0.99*df_kisyo['平均気温(℃)'] - 14.3) + 46.3
df_kisyo['WCI'] = (33-df_kisyo['平均気温(℃)'])*(10.45+10*(df_kisyo['平均風速(m/s)']**0.5)-df_kisyo['平均風速(m/s)'])
df_kisyo = df_kisyo.rename(columns={'年月日': 'datetime'})
df_kisyo['datetime'] = pd.to_datetime(df_kisyo['datetime'])
df_kisyo =df_kisyo.set_index('datetime')
```

```python
df_kisyo.head()
```

```python
#WBGTデータ
df_wbgt = pd.read_csv(processed_data_path+'WBGT_Tokyo.csv')
df_wbgt = df_wbgt.rename(columns={'Date': 'datetime'})
df_wbgt['datetime'] = pd.to_datetime(df_wbgt['datetime'])
df_wbgt = df_wbgt[(pd.to_datetime(df_wbgt['datetime']) >= cri_1)]
df_wbgt = df_wbgt[(pd.to_datetime(df_wbgt['datetime']) <= cri_2)]
df_wbgt = df_wbgt.set_index('datetime')
```

# 相関係数

```python
from scipy import stats
```

```python
correlation, pvalue = stats.pearsonr(df_stroke['搬送人員（計）'], df_wbgt['WBGT'])
print(correlation,pvalue)
```

```python
correlation, pvalue = stats.pearsonr(df_stroke['搬送人員（計）'], df_apparent['体感温度'])
print(correlation,pvalue)
```

```python
correlation, pvalue = stats.pearsonr(df_stroke['搬送人員（計）'], df_kisyo['平均気温(℃)'])
print(correlation,pvalue)
```

# 順位相関係数

```python
#順位相関係数
correlation, pvalue = stats.spearmanr(df_stroke['搬送人員（計）'], df_wbgt['WBGT'])
print(correlation,pvalue)
```

```python
correlation, pvalue = stats.spearmanr(df_stroke['搬送人員（計）'], df_apparent['体感温度'])
print(correlation,pvalue)
```

```python
correlation, pvalue = stats.spearmanr(df_stroke['搬送人員（計）'], df_kisyo['平均気温(℃)'])
print(correlation,pvalue)
```
